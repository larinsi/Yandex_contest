### G. Наибольшее произведение двух чисел
* Ограничение времени: 1 секунда
* Ограничение памяти: 64Mb
* Ввод: стандартный ввод или input.txt
* Вывод: стандартный вывод или output.txt

Дан список, заполненный произвольными целыми числами. Найдите в этом списке два числа, произведение которых максимально. Выведите эти числа в порядке неубывания.

Список содержит не менее двух элементов. Числа подобраны так, что ответ однозначен.

Решение должно иметь сложность O(n), где n - размер списка. 

Пример 1:

Ввод:
```

```
Вывод:
```
5 5

```

````python
def max_multiplication(array: list):
    max2, max1 = array[0], array[1]
    min1, min2 = array[0], array[1]
    if max2 > max1:
        max2, max1 = max1, max2
        min1, min2 = min2, min1
    for i in range(2, len(array)):
        if array[i] >= max1:
            max2, max1 = max1, array[i]
        elif array[i] >= max2:
            max2 = array[i]
        if array[i] <= min1:
            min1, min2 = array[i], min1
        elif array[i] <= min2:
            min2 = array[i]
    if (min1 * min2) > (max1 * max2):
        result = (min1, min2)
    else:
        result = (max2, max1)
    return result
````

````python
array = [int(i) for i in input().split(" ")]
````

````python
max_multiplication([-1, -2])
````
